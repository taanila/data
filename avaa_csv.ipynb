<h1>csv-datan avaaminen</h1>

- Pandas kirjaston <b>read_csv()</b> avaa csv-muotoisen (comma separated value) tiedoston.

- Pelkällä tiedostonimellä avaat koodin kanssa samaan kansioon tallennetun csv-tiedoston.

- Jos tiedosto on muussa kansiossa tai netissä, niin lisää tiedostopolku tai nettiosoite, esimerkiksi <i>'C:/Users/Aki/Documents/data1.csv'</i>.

- Suomessa csv-tiedoston erottimena käytetään usein puolipistettä pilkun sijasta, jotta pilkkua voidaan käyttää desimaalierottimena. Tällöin tarvitset parametreja <b>sep = ';'</b> ja <b>decimal = ','</b>.  

- Datan alkuosaa voit avaamisen jälkeen katsoa komennolla <b>df.head()</b> ja loppuosaa komennolla <b>df.tail()</b>.

```python
import pandas as pd
```

```python
# Suomalaisittain tallennetun tiedoston avaamisessa tarvitsen sep- ja decimal-parametreja
df = pd.read_csv('https://taanila.fi/data1.csv', sep = ';', decimal = ',')
df.head()
```

```python
df.tail()
```

Jos erotinmerkki (sep) ja desimaalierotin (decimal) eivät täsmää tiedoston kanssa tai jos tiedoston alussa on ylimääräistä selittävää tekstiä, niin read_csv antaa virheilmoituksen. Tällaisessa tapauksessa kannattaa nuuskia etukäteen tiedoston sisältöä muilla keinoin.

Voit avata tiedoston tekstieditoriin, esimerkiksi WordPadiin. Windowsin mukana tuleva WordPad on hyvä vaihtoehto, koska se avaa isotkin tiedostot toisin kuin Notepad.

Jos tiedosto on netissä, niin nuuskiminen Pythonista käsin onnistuu esimerkiksi seuraavasti:

```python
import requests
tiedosto = requests.get('https://taanila.fi/elisa.csv')
print(tiedosto.text)
```

Edellisen perusteella avaaminen tietokehikkoon onnistuu seuraavasti (skiprows = 1 hyppää ensimmäisen rivin yli):

```python
df = pd.read_csv('https://taanila.fi/elisa.csv', sep = ';', decimal = ',', skiprows = 1)
df.head()
```
